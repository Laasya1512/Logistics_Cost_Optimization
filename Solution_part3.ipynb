```python
import pandas as pd
df=pd.read_csv('Comparison_Data - Sheet1.csv')
df
```

```python
import matplotlib.pyplot as plt
plt.plot(df['Indices'],df['Railways'],label='Railways')
plt.plot(df['Indices'],df['Roads'],label='Roads')
plt.plot(df['Indices'],df['Coastal'],label='Coastal')
plt.xlabel('Indices')
plt.ylabel('Modes of Transport')
plt.title('Characteristics of different Transportation')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
import matplotlib.pyplot as plt
labels = 'Railways', 'Road', 'Coastal'
sizes = [52, 31, 17]
explode = (0, 0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()#This is with respect to Capacity(tons/day)
```
