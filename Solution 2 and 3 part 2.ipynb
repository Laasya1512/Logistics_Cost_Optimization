## Formatting the maximum and minimum value of every attribute

```python
format_dict ={'Road_Total_Cost':'Rs.{0:,.2f}'} 
#Simplified format dictionary with values that do make sense for our data
df.head(10).style.format(format_dict).highlight_max(color='lightgreen').highlight_min(color='skyblue')
df.head(10).style.format(format_dict).highlight_max(color='lightgreen').highlight_min(color='skyblue')
```

## Line plot comparing the total cost of the three modes of shipping

```python
import matplotlib.pyplot as plt
plt.plot(df['Route_Number'], df['Road_Total_Cost'], label='Road',color='purple')
plt.plot(df['Route_Number'], df['Rail_Total_Cost'], label='Rail',color='teal')
plt.plot(df['Route_Number'], df['Sea_Total_Cost'], label='Sea-Road',color='blue')
#label for x-axis
plt.ylabel('Routes')
#label for y-axis
plt.title('Transport Medium')
#label for the title of the graph
plt.grid('True')
#for including grid
plt.legend()
#for the legend
```

## Line plot comparing the total time for the three modes of shipping

```python
plt.plot(df['Route_Number'], df['Road_Transit_time(hours)'], label='Road',color='purple')
plt.plot(df['Route_Number'], df['Rail_Transit_time_(hours)'], label='Rail',color='teal')
plt.plot(df['Route_Number'], df['SeaRoad_Transit_time_(hours)'], label='Sea-Road',color='blue')
#label for x-axis
plt.ylabel('Route')
#label for y-axis
plt.title('Transport Time')
#label for the title of the graph
plt.grid('True')
#for including grid
plt.legend()
#for the legend
```

## Sub-plots of Total cost of each mode of shipping
